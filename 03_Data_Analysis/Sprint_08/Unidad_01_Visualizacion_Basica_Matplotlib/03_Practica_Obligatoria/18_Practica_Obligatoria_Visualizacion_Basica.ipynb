### Ejercicio 1

#### #1.1

- Carga el dataset de casas de California que est en la ruta "./data/california_cities.csv".  
- Muestra parte de su contenido y su descripcin.  
- Deshazte de las filas con nulos.

```python
df = pd.read_csv("data/california_cities.csv", index_col= 0)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["elevation_m"] = df["elevation_m"].fillna(df["elevation_m"].mean())
df["elevation_ft"] = df["elevation_ft"].fillna(df["elevation_ft"].mean())
df["area_total_sq_mi"] = df["area_total_sq_mi"].fillna(df["area_total_sq_mi"].mean())
df["area_water_sq_mi"] = df["area_water_sq_mi"].fillna(df["area_water_sq_mi"].mean())
df["area_total_km2"] = df["area_total_km2"].fillna(df["area_total_km2"].mean())
df["area_land_km2"] = df["area_land_km2"].fillna(df["area_land_km2"].mean())
df["area_water_km2"] = df["area_water_km2"].fillna(df["area_water_km2"].mean())
df["area_water_percent"] = df["area_water_percent"].fillna(df["area_water_percent"].mean())
df.isna().sum()
```

#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center" o investigar como funciona bar_label). Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrn.  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Arial", tamao = 14, peso = "Bold", color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.
4. El boxplot de la altura de las ciudades.

```python
#1. 
top5 = df.sort_values("population_total", ascending=False).head(5)

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(
    top5["city"],
    top5["population_total"],
    color=["blue", "gray", "red", "olive", "brown"],
    alpha=0.5
)

ax.bar_label(bars, labels=[f"{int(v):,}" for v in top5["population_total"]])

ax.set_title("Las ciudades ms pobladas de California")
ax.set_ylabel("Poblacin")
ax.set_xlabel("Ciudad")

plt.show()

```

```python
#2. 

plt.figure(figsize=(10, 6))

plt.hist(df["area_total_km2"], bins=50, color="steelblue")

plt.title(
    "Histogrma del rea total ocupada por las ciudades",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

plt.xlabel(
    "Superficie (km)",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

plt.ylabel(
    "Nmero de ciudades",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

plt.xticks(fontname="Arial", fontsize=14, fontweight="bold")
plt.yticks(fontname="Arial", fontsize=14, fontweight="bold")

plt.show()
```

```python
#3.

plt.figure(figsize=(10, 6))

plt.scatter(
    df["area_water_km2"],
    df["population_total"],
    color="green",
    alpha=0.5
)

outliers = df.sort_values("population_total", ascending=False).head(2)

for _, row in outliers.iterrows():
    plt.annotate(
        "Outliers",
        xy=(row["area_water_km2"], row["population_total"]),
        xytext=(row["area_water_km2"] * 0.4, row["population_total"] * 0.7),
        arrowprops=dict(arrowstyle="->", color="black"),
        ha="center"
    )

plt.title("Poblacin por kilmetro cuadrado de agua")
plt.xlabel("rea de agua (km)")
plt.ylabel("Poblacin")

plt.show()
```

```python
#4.

plt.figure(figsize=(8, 6))

plt.boxplot(df["elevation_m"], vert=False, patch_artist=True, boxprops=dict(facecolor="red"))

plt.title("Boxplot de la altitud de las ciudades de California")
plt.ylabel("Altura (metros)")

plt.show()
```

```python
#. Todo junto.

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

top5 = df.sort_values("population_total", ascending=False).head(5)

bars = axes[0, 0].bar(
    top5["city"],
    top5["population_total"],
    color=["blue", "gray", "red", "olive", "brown"],
    alpha=0.5
)

axes[0, 0].bar_label(bars, labels=[f"{int(v):,}" for v in top5["population_total"]])

axes[0, 0].set_title("Las ciudades ms pobladas de California")
axes[0, 0].set_ylabel("Poblacin")
axes[0, 0].set_xlabel("Ciudad")


axes[0, 1].hist(df["area_total_km2"], bins=50, color="steelblue")

axes[0, 1].set_title(
    "Histogrma del rea total ocupada por las ciudades",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

axes[0, 1].set_xlabel(
    "Superficie (km)",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

axes[0, 1].set_ylabel(
    "Nmero de ciudades",
    fontname="Arial",
    fontsize=14,
    fontweight="bold"
)

for tick in axes[0, 1].get_xticklabels():
    tick.set_fontname("Arial")
    tick.set_fontsize(14)
    tick.set_fontweight("bold")

for tick in axes[0, 1].get_yticklabels():
    tick.set_fontname("Arial")
    tick.set_fontsize(14)
    tick.set_fontweight("bold")


axes[1, 0].scatter(
    df["area_water_km2"],
    df["population_total"],
    color="green",
    alpha=0.5
)

outliers = df.sort_values("population_total", ascending=False).head(2)

for _, row in outliers.iterrows():
    axes[1, 0].annotate(
        "Outliers",
        xy=(row["area_water_km2"], row["population_total"]),
        xytext=(row["area_water_km2"] * 0.4, row["population_total"] * 0.7),
        arrowprops=dict(arrowstyle="->", color="black"),
        ha="center"
    )

axes[1, 0].set_title("Poblacin por kilmetro cuadrado de agua")
axes[1, 0].set_xlabel("rea de agua (km)")
axes[1, 0].set_ylabel("Poblacin")


axes[1, 1].boxplot(
    df["elevation_m"],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor="red")
)

axes[1, 1].set_title("Boxplot de la altitud de las ciudades de California")
axes[1, 1].set_ylabel("Altura (metros)")

plt.tight_layout()
plt.show()
```
