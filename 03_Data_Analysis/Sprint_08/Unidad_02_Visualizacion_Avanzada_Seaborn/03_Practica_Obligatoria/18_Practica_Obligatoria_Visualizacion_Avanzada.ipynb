### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python

def scatter_plot(df, x, y, cat=None, tam=40, escala=1.0):
    # Si tam es una columna, usamos sus valores como tamao (con escala)
    if isinstance(tam, str):
        sizes = df[tam] * escala
    else:
        sizes = tam

    plt.figure(figsize=(10, 6))

    # Caso sin variable categrica
    if cat is None:
        plt.scatter(df[x], df[y], s=sizes)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.title(f"{y} vs {x}")
        plt.show()
        return

    # Caso con variable categrica
    categorias = df[cat].astype("category").cat.categories
    colores = sns.color_palette(n_colors=len(categorias))

    for c, col in zip(categorias, colores):
        mask = df[cat] == c
        plt.scatter(
            df.loc[mask, x],
            df.loc[mask, y],
            s=sizes[mask] if isinstance(tam, str) else sizes,
            color=col,
            label=str(c)
        )

    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"{y} vs {x}")
    plt.legend(title=cat)
    plt.show()

```

```python
scatter_plot(
    df_california,
    x="area_total_km2",
    y="population_total",
    tam="population_total",
    escala=1/10000
)
```

```python
df_titanic = pd.read_csv("./data/titanic.csv")
```

```python
df_titanic.info()
```

```python
conteo = df_titanic["survived"].value_counts().sort_index()

plt.figure(figsize=(6, 4))
plt.bar(
    ["Fallecidos", "Supervivientes"],
    conteo,
    alpha=0.7
)

plt.title("Tasa de mortalidad en el Titanic")
plt.ylabel("Nmero de pasajeros")
plt.show()
```

```python
supervivencia_sexo = (
    df_titanic
    .groupby("sex")["survived"]
    .mean()
)

plt.figure(figsize=(6, 4))
plt.bar(
    supervivencia_sexo.index,
    supervivencia_sexo,
    alpha=0.7
)

plt.title("Tasa de supervivencia por sexo")
plt.ylabel("Proporcin de supervivientes")
plt.ylim(0, 1)
plt.show()
```

```python
supervivencia_clase = (
    df_titanic
    .groupby("pclass")["survived"]
    .mean()
    .sort_index()
)

plt.figure(figsize=(6, 4))
plt.bar(
    supervivencia_clase.index.astype(str),
    supervivencia_clase,
    alpha=0.7
)

plt.title("Tasa de supervivencia por clase")
plt.xlabel("Clase")
plt.ylabel("Proporcin de supervivientes")
plt.ylim(0, 1)
```

```python
supervivencia_sexo_clase = (
    df_titanic
    .groupby(["sex", "pclass"])["survived"]
    .mean()
    .unstack()
)

supervivencia_sexo_clase.plot(
    kind="bar",
    figsize=(8, 5),
    alpha=0.7
)

plt.title("Tasa de supervivencia por sexo y clase")
plt.ylabel("Proporcin de supervivientes")
plt.xlabel("Sexo")
plt.ylim(0, 1)
plt.legend(title="Clase")
plt.show()
```

```python
tabla = (
    df_titanic
    .dropna(subset=["embarked", "pclass"])
    .groupby(["embarked", "pclass"])["survived"]
    .mean()
    .unstack()
    .sort_index()
)

tabla.plot(kind="bar", figsize=(8, 5), alpha=0.7)

plt.title("Tasa de supervivencia por puerto y clase")
plt.xlabel("Puerto de embarque")
plt.ylabel("Proporcin de supervivientes")
plt.ylim(0, 1)
plt.legend(title="Clase")
plt.show()
```
